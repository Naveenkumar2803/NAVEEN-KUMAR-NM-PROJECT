```python
traveller_types = pd.get_dummies(df['Type_of_Travellers'], drop_first=True)
traveller_types
```

```python
df_transform = df.merge(traveller_types, left_on=df.index, right_on=traveller_types.index)
df_transform.drop(['key_0', 'Type_of_Travellers'], axis=1, inplace=True)
df_transform
```

```python
seat_types = pd.get_dummies(df_transform['Seat_Types'], drop_first=True)
seat_types
```

```python
df_transform = df_transform.merge(seat_types, left_on=df_transform.index, right_on=seat_types.index)
df_transform.drop(['key_0', 'Seat_Types'], axis=1, inplace=True)
df_transform
```

```python
aircraft_encoder = ce.BinaryEncoder(cols=['Aircraft Type'], return_df=True)
df_transform = aircraft_encoder.fit_transform(df_transform)
df_transform
```

```python
country_encoder = ce.BinaryEncoder(cols=['Country'], return_df=True)
df_transform = country_encoder.fit_transform(df_transform)
df_transform
```

```python
df_transform.drop(['Wifi & Connectivity', 'Inflight Entertainment'], axis=1, inplace=True)
df_transform
```

```python
df_transform['Seat Comfort'] = df_transform['Seat Comfort'].fillna(df_transform['Seat Comfort'].mode()[0])
df_transform['Cabin Staff Service'] = df_transform['Cabin Staff Service'].fillna(df_transform['Cabin Staff Service'].mode()[0])
df_transform['Ground Service'] = df_transform['Ground Service'].fillna(df_transform['Ground Service'].mode()[0])
df_transform['Food & Beverages'] = df_transform['Food & Beverages'].fillna(df_transform['Food & Beverages'].mode()[0])
df_transform
```

```python
df_transform.isna().any()
```

```python
df_transform['Seat Comfort'] = df_transform['Seat Comfort'].astype("int64")
df_transform['Cabin Staff Service'] = df_transform['Cabin Staff Service'].astype("int64")
df_transform['Ground Service'] = df_transform['Ground Service'].astype("int64")
df_transform['Food & Beverages'] = df_transform['Food & Beverages'].astype("int64")
df_transform
```

```python
df_transform['Route'] = df_transform['Route'].fillna('')
df_transform[['Departure', 'PartTwo']] = df_transform['Route'].str.split(' to ', n=1, expand=True)
df_transform['PartTwo'] = df_transform['PartTwo'].fillna('')
df_transform[['Destination', 'Via']] = df_transform['PartTwo'].str.split(' via ', n=1, expand=True)
df_transform.drop(['PartTwo'], axis=1, inplace=True)
df_transform['Via'] = df_transform['Via'].fillna('')
df_transform
```
